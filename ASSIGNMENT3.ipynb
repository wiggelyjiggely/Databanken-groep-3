### Query 1

Schrijf een query met als output:

Een dataframe dat de displaynames van alle users bevat, oplopend gesorteerd volgens displayname.

```python
def get_displaynames_of_users(connection):
    # Actual query
    query = "SELECT DisplayName FROM users ORDER BY DisplayName"

    # EXTRA CODE HIER
    
    # Making a cursor and executing the query
    cursor=connection.cursor()
    cursor.execute(query)
    
    # Collecting the result and casting it in a pd.DataFrame
    res= cursor.fetchall()
    df=pd.DataFrame(res,columns=['DisplayName'])
    
    return df
```

leeftijd 'age', oplopend gesorteerd volgens leeftijd

### Query 3

Schrijf een query met als output:

Een dataframe dat alle attributen bevat van de User relatie, zodat enkel users voorkomen die na 'date' (ook strikt) nog een post hebben aangemaakt (**en dus 'owner' zijn**), oplopend gesorteerd volgens displayname. Zorg er voor dat deze lijst geen dubbels bevat!

```python
def get_users_with_recent_post(connection,date):
    # Actual query
    query="""SELECT DISTINCT * 
            FROM users 
            WHERE 
            EXISTS (SELECT * 
            FROM posts
            WHERE users.Id = posts.OwnerUserId
            AND DATE(posts.CreaionDate) >= %s)
            ORDER BY users.DisplayName ASC""" % (date)
    
    # EXTRA CODE HIER

    # Making a cursor and executing the query
    cursor=connection.cursor()
    cursor.execute(query)
    
    # Collecting the result and casting it in a pd.DataFrame
    res= cursor.fetchall()
    
    """
    De DataFrame krijgt hier reeds expliciet de namen mee van alle kolommen van de user relatie. 
    Als je query meer of minder kolommen teruggeeft, 'past' deze niet in de DataFrame en zal je een error 
    krijgen, (zelfs als de query op zich wel gelukt is). 
    
    Houd dus rekening met dit gedrag als je je query aan het testen bent. In de testfase kan je eventueel het 'columns'
    argument weglaten om je tussenresultaten te bekijken. Vergeet echter niet om in je finale oplossing terug de door
    ons gespecifieerde kolomnamen toe te voegen, anders wordt de oplossing als helemaal fout aangerekend.
    
    Deze opmerking geldt voor alle volgende queries.
    """
    df=pd.DataFrame(res,columns=user_columns)
    
    
    return df
```

### Query 5

Schrijf een query met als output:

Een dataframe dat alle attributen bevat van de Post relatie, zodat enkel posts voorkomen waar **minstens 1** user met minstens reputatie 'reputation' op gevote heeft, oplopend geordend volgens post Id.

```python
def get_posts_at_least_one_high_reputation_vote(connection,reputation):
    # Actual query
    query="""SELECT * 
            FROM posts
            WHERE 
            EXISTS(
                SELECT *
                FROM users JOIN votes ON users.Id = votes.UserId
                WHERE users.Reputation >= %i
                AND posts.Id = votes.PostId)
            ORDER BY posts.Id""" % (reputation)
    
    # EXTRA CODE HIER

    # Making a cursor and executing the query
    cursor=connection.cursor()
    cursor.execute(query)
    
    # Collecting the result and casting it in a pd.DataFrame
    res= cursor.fetchall()
    df=pd.DataFrame(res,columns=post_columns)
    return df
```

### Query 10
